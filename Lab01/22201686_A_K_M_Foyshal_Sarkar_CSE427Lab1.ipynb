Use NumPy library to complete the following tasks:
1. Create two linear arrays a and b with 20 random elements each ranging from 1 to 100.
2. Now change those linear arrays to two 5x4 matrices.
3. Perform scalar multiplication with matrix a and b.
4. Now try to modify the matrix b and perform matrix multiplication with a and b to store it in
c.
5. Print only the column 2, 3 of matrix c. Print the whole matrix using slicing technique.
6. Find the largest element in the array, and also the index of that element.
7. Convert the matrix c into an one dimensional array keeping the original array
unchanged.


```python
#Task 1

import numpy as np


a = np.random.randint(1, 101, 20)
b = np.random.randint(1, 101, 20)

print("Linear array a:\n", a)
print("\nLinear array b:\n", b)
```

```python
#task 2

a = a.reshape(5, 4)
b = b.reshape(5, 4)

print("\nMatrix a (5x4):\n", a)
print("\nMatrix b (5x4):\n", b)
```

```python
#task 3

scalar = 5
a_scaled = a * scalar
b_scaled = b * scalar

print("Matrix A after scalar multiplication:\n", a_scaled)
print("\nMatrix B after scalar multiplication:\n", b_scaled)
```

```python
#task 4
b_modified = b.T
c = np.matmul(a, b_modified)
print("\nModified matrix b (transposed):\n", b_modified)
print("\nMatrix multiplication result c = a * b_modified:\n", c)
```

```python
#task 5
print("\nColumns 2 and 3 of matrix c:\n", c[:, 1:3])
print("\nWhole matrix c using slicing:\n", c[:, :])
```

```python
#task 6
max_value = np.max(c)                      # largest value in the whole matrix
flat_index = np.argmax(c)                  # index in flattened (1D) form
row_col_index = np.unravel_index(flat_index, c.shape)  # (row, column)

print("\nLargest element in matrix c:", max_value)
print("Flat index of largest element:", flat_index)
print("Row–Column index of largest element:", row_col_index)
print(f"\nLargest element in c: {max_value} at position {max_index}")
# # 6. Find the largest element in c and its index.
# max_value = np.max(c)  # largest element in matrix 'c'
# max_index_raw = np.unravel_index(np.argmax(c), c.shape)  # Finds the index position (as NumPy ints)
# max_index = (int(max_index_raw[0]), int(max_index_raw[1]))  # Converting both row and column index to normal Python int

# print(f"\nLargest element in c: {max_value} at position {max_index}")
```

```python
#task 7

c_flat = c.flatten()
print("\nFlattened (1D) version of matrix c:\n", c_flat)
```

```python
#Task 1

import pandas as pd
url = "https://drive.google.com/uc?id=1zo5y462G1FaGUlAEUayACHSV2zSliR-5"
df = pd.read_csv(url)
print(df)
```

```python
#Task2

print("---- First Few Rows of the Dataset ----")
print(df.head(10), "\n")
```

```python
#task 3

df.describe(include='all')
```

```python
#task 4
df.isnull().sum()
```

```python
#task 5

print("Before replacing null values:\n")
print(df.isnull().sum(), "\n")

# Replace null values in numeric columns with the mean of that column
df.fillna(df.mean(numeric_only=True), inplace=True)

# Show how many null values are present after replacement
print("After replacing null values with column mean:\n")
print(df.isnull().sum())
```

```python
#TAsk 1
import matplotlib.pyplot as plt
temperature_dhaka = [25, 34, 21, 45, 28, 6, 43, 18, 7, 2]
humidity_dhaka = [28, 25, 29, 20, 26, 50, 19, 29, 52, 55]
plt.scatter(temperature_dhaka, humidity_dhaka, color='red', marker='*')

# Add labels and title
plt.title("Temperature vs Humidity in Dhaka")
plt.xlabel("Temperature (°C)")
plt.ylabel("Humidity (%)")

# Show the graph
plt.show()
```

```python
#task 2
import matplotlib.pyplot as plt

# Given data
study_hours = [2, 3, 4, 4, 5, 6, 7, 7, 8, 9, 9, 10, 11, 11, 12]
marks = [6, 10, 15, 20, 34, 44, 55, 60, 55, 67, 70, 80, 90, 99, 100]

# Set the figure size
plt.figure(figsize=(12, 8))

# Create a line plot
plt.plot(study_hours, marks, color='blue', marker='o', linewidth=2)

# Add labels, title, and grid
plt.title("Study Hours vs Marks", fontsize=16)
plt.xlabel("Study Hours", fontsize=14)
plt.ylabel("Marks", fontsize=14)
plt.grid(True)

# Display the plot
plt.show()
```

```python
#task 3
# Given data
subjects = ['Maths', 'English', 'Science', 'Physics', 'Computer']
marks = [89, 90, 45, 78, 99]

# Create a figure for subplots
plt.figure(figsize=(12, 8))

# -------------------------------
# Task 3: Horizontal Bar Graph
# -------------------------------
plt.subplot(1, 2, 1)  # (rows, columns, position)
plt.barh(subjects, marks, color='orange')
plt.title("Horizontal Bar Graph: Subject Scores")
plt.xlabel("Marks")
plt.ylabel("Subjects")
```

```python
#task 4
colors = ['red', 'blue', 'green', 'purple', 'cyan']
plt.figure(figsize=(12, 8))
plt.subplot(1, 2, 2)
plt.bar(subjects, marks, color=colors)
plt.title("Vertical Bar Graph: Subject Scores")
plt.xlabel("Subjects")
plt.ylabel("Marks")
```

```python
#task 5
plt.figure(figsize=(12, 10))

# 1️⃣ Scatter plot: temperature vs humidity
plt.subplot(2, 2, 1)
plt.scatter(temperature_dhaka, humidity_dhaka, color='red', marker='*')
plt.title("Temperature vs Humidity (Dhaka)")
plt.xlabel("Temperature (°C)")
plt.ylabel("Humidity (%)")

# 2️⃣ Line plot: study hours vs marks
plt.subplot(2, 2, 2)
# Use the 'marks' variable from task 2 for the line plot
study_hours = [2, 3, 4, 4, 5, 6, 7, 7, 8, 9, 9, 10, 11, 11, 12]
marks_study = [6, 10, 15, 20, 34, 44, 55, 60, 55, 67, 70, 80, 90, 99, 100]
plt.plot(study_hours, marks_study, color='blue', marker='o')
plt.title("Study Hours vs Marks")
plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.grid(True)

# 3️⃣ Horizontal bar chart
plt.subplot(2, 2, 3)
# Use the 'subjects' and 'marks' variables from task 3/4 for the bar charts
subjects_bar = ['Maths', 'English', 'Science', 'Physics', 'Computer']
marks_bar = [89, 90, 45, 78, 99]
plt.barh(subjects_bar, marks_bar, color='orange')
plt.title("Subject-wise Marks (Horizontal)")
plt.xlabel("Marks")
plt.ylabel("Subjects")

# 4️⃣ Vertical bar chart with different colors
colors = ['red', 'blue', 'green', 'purple', 'cyan']
plt.subplot(2, 2, 4)
plt.bar(subjects_bar, marks_bar, color=colors)
plt.title("Subject-wise Marks (Vertical)")
plt.xlabel("Subjects")
plt.ylabel("Marks")

# Adjust layout and show all together
plt.tight_layout()
plt.show()
```
